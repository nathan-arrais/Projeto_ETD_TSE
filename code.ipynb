Atividade 04 – Após a criação da lista e do dicionário, criar uma rotina para ler o diretório raiz e realizar a movimentação dos arquivos conforme encontra-se no dicionário. 

```python
import shutil 
for item in lista:
  if item.endswith('pdf'):
    os.makedirs(r'C:\Users\natha\OneDrive\Documentos\MBA_ENG_DADOS\ETD\PDF', exist_ok=True)
    shutil.move(r'C:\Users\natha\OneDrive\Documentos\MBA_ENG_DADOS\ETD' + '\\' + item, r'C:\Users\natha\OneDrive\Documentos\MBA_ENG_DADOS\ETD\PDF')
    print(item)

for item in lista:
  if item.endswith('csv'):
    os.makedirs(r'C:\Users\natha\OneDrive\Documentos\MBA_ENG_DADOS\ETD\CSV', exist_ok=True)
    shutil.move(r'C:\Users\natha\OneDrive\Documentos\MBA_ENG_DADOS\ETD' + '\\' + item, r'C:\Users\natha\OneDrive\Documentos\MBA_ENG_DADOS\ETD\CSV')
    print(item)
```

Atividade 05 – Depois de movimentar os arquivos para os diretórios corretos, eles precisam ser organizados por UF. Caso o diretório da UF não exista, o mesmo deverá ser criado em tempo de execução. 

```python
diretorio_csv = fr'C:\Users\natha\OneDrive\Documentos\MBA_ENG_DADOS\ETD\CSV'

for item in lista:  
  for uf in list_uf:
    if f'_{uf}' in item:  
      destino_dir = os.path.join(diretorio_csv, uf)
      os.makedirs(destino_dir, exist_ok=True)
      origem = os.path.join(diretorio_csv, item)
      destino = os.path.join(destino_dir, item)
      shutil.move(origem, destino)
      break
```

Atividade 06 – Após “arrumar” o ambiente de trabalho, verificar se restou algum arquivo no diretório raiz (local onde os arquivos foram descompactados). Caso exista, na lista de UF, deverá ser inserido um elemento com o nome “Outro” e no dicionário Diretório também é necessário realizar a inclusão de um diretório para que ele seja movimentado.

```python
resto = os.listdir(diretorio_raiz)
arquivos_restantes = [arquivo for arquivo in resto if os.path.isfile(os.path.join(diretorio_raiz, arquivo))]

if arquivos_restantes:
  if "Outros" not in list_uf:
    list_uf.append("Outros")

  outros_dir = os.path.join(diretorio_raiz, "Outros")
  os.makedirs(outros_dir, exist_ok=True)
  
  for arquivo_restante in arquivos_restantes:
    origem = os.path.join(diretorio_raiz, arquivo_restante)
    destino = os.path.join(outros_dir, arquivo_restante)

  shutil.move(origem, destino)
  print(f"Movendo {arquivo_restante} para o diretório 'Outros'")
```
